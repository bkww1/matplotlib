```python
import numpy as np
```

```python
#Załadowanie danych z plików do zmiennych data
data_1c = np.genfromtxt('1c.csv', delimiter=',')
data_1crs = np.genfromtxt('1crs.csv', delimiter=',')
data_1ers = np.genfromtxt('1ers.csv', delimiter=',')
data_2c = np.genfromtxt('2c.csv', delimiter=',')
data_2crs = np.genfromtxt('2crs.csv', delimiter=',')
```

```python
#Załadowanie wwszystkich wartości od wiersza o indekszie 1 do końca i kolumny o indeksie 2 do końca
data_1c_val = data_1c[1:,2:]
data_1crs_val = data_1crs[1:,2:]
data_1ers_val = data_1ers[1:,2:]
data_2c_val = data_2c[1:,2:]
data_2crs_val = data_2crs[1:,2:]
```

```python
#Utworzenie tablicy ze srednich wartosci dla kazdego zbioru danych
data_val_avg =[]
data_val_avg.append(np.sum(data_1c_val, axis=1) / 32)
data_val_avg.append(np.sum(data_1crs_val, axis=1) /32)
data_val_avg.append(np.sum(data_1ers_val, axis=1) /32)
data_val_avg.append(np.sum(data_2c_val, axis=1) /32)
data_val_avg.append(np.sum(data_2crs_val, axis=1) /32)
```

```python
#Array wartości dla osi X
data_xaxis = data_1c[1:,1]
data_xaxis = np.array(data_xaxis)
```

```python
from matplotlib import pyplot as plt
#Wykonujemy plot dla poszczególnych danych
plt.figure(figsize=(5, 5))
plt.plot(data_xaxis, data_val_avg[2], label=f'1-Evol-RS', color='blue', linewidth = 1)
plt.plot(data_xaxis, data_val_avg[1], label=f'1-Coev-RS', color='green', linewidth = 1)
plt.plot(data_xaxis, data_val_avg[4], label=f'2-Coev-RS', color='red', linewidth = 1)
plt.plot(data_xaxis, data_val_avg[0], label=f'1-Coev', color='black', linewidth = 1)
plt.plot(data_xaxis, data_val_avg[3], label=f'2-Coev', color='violet', linewidth = 1)

#Ustawienia osi
plt.xlabel('Rozegranych gier')
plt.xlim(0,500000)
plt.ylabel('Odsetek wygranych gier')
plt.ylim(0.6,1)
plt.legend(loc='lower right', bbox_to_anchor=(1.0, 0.0), fontsize=12)
plt.tick_params(axis='both', direction='in', top=True, right=True)


plt.savefig('myplot1.pdf')
plt.show()
```

```python
data_val_avg = []
data_val_avg.append((np.sum(data_1c_val, axis=1) / 32) * 100)
data_val_avg.append((np.sum(data_1crs_val, axis=1) / 32) * 100)
data_val_avg.append((np.sum(data_1ers_val, axis=1) / 32) * 100)
data_val_avg.append((np.sum(data_2c_val, axis=1) / 32) * 100)
data_val_avg.append((np.sum(data_2crs_val, axis=1) / 32) * 100)
# Array wartości dla osi X
data_xaxis = data_1c[1:, 1]
data_xaxis = np.array(data_xaxis) / 1000
```

```python
from matplotlib import pyplot as plt
import matplotlib as mpl

# Ustawienia markerów
top_axis_marker_ticks = slice(0, 200, 25)

# Ogólne ustawienia wszystkich plotów
mpl.rcParams["lines.markeredgecolor"] = "black"
mpl.rcParams["lines.markeredgewidth"] = 0.5
mpl.rcParams["lines.linewidth"] = 1

# Wykonujemy plot dla poszczególnych danych
plt.figure(figsize=(3.5, 8))
plt.plot(
    data_xaxis,
    data_val_avg[2],
    label=f"1-Evol-RS",
    color="blue",
    marker=".",
    markevery=top_axis_marker_ticks,
    markersize=10,
)
plt.plot(
    data_xaxis,
    data_val_avg[1],
    label=f"1-Coev-RS",
    color="green",
    marker="v",
    markevery=top_axis_marker_ticks,
    markersize=7,
)
plt.plot(
    data_xaxis,
    data_val_avg[4],
    label=f"2-Coev-RS",
    color="red",
    marker="d",
    markevery=top_axis_marker_ticks,
    markersize=6,
)
plt.plot(
    data_xaxis,
    data_val_avg[0],
    label=f"1-Coev",
    color="black",
    marker="s",
    markevery=top_axis_marker_ticks,
    markersize=6,
)
plt.plot(
    data_xaxis,
    data_val_avg[3],
    label=f"2-Coev",
    color="violet",
    marker="d",
    markevery=top_axis_marker_ticks,
    markersize=6,
)

# Ustawienia osi
plt.xlabel(r"Rozegranych gier ($\times$1000)")
plt.xlim(0, 500)
plt.ylabel("Odsetek wygranych gier [%]")
plt.ylim(60, 100)
plt.tick_params(axis="both", direction="in", top=True, right=True)
plt.grid(color="black", linestyle="--", linewidth=0.3, dashes=(3, 20))

# Ustawienia legendy
plt.legend(loc="lower right", bbox_to_anchor=(1.0, 0.0), fontsize=12)
plt.legend(numpoints=2, handlelength=2.0)

# Górna oś Pokoleń
top_axis = plt.twiny()
top_axis.set_xlabel("Pokolenie")
top_axis.set_xlim(0, 200)
top_axis_ticks = np.arange(0, 201, 40)
top_axis.set_xticks(top_axis_ticks)


plt.savefig("myplot2.pdf")
plt.show()
```

```python
data_bp = [
    data_1ers[200, 2:] * 100,
    data_1crs[200, 2:] * 100,
    data_2crs[200, 2:] * 100,
    data_1c[200, 2:] * 100,
    data_2c[200, 2:] * 100,
]

flier_style = dict(
    marker="+",
    markerfacecolor="b",
    markeredgecolor="b",
    markeredgewidth=0.8,
    markersize=10,
)
median_style = dict(color="r", linewidth=2)
box_style = dict(color="b", linewidth=1.5)
whisker_style = dict(color="b", linewidth=1.5, linestyle="--", dashes=(7, 5.3))
means_style = dict(
    marker="o",
    markerfacecolor="b",
    markeredgecolor="black",
    markeredgewidth=2,
    markersize=6,
)
caps_style = dict(linewidth=1.5)

plt.figure(figsize=(3.5, 8))
plt.boxplot(
    data_bp,
    labels=["1-Evol-RS", "1-Coev-RS", "2-Coev-RS", "1-Coev", "2-Coev"],
    capprops=caps_style,
    showmeans=True,
    meanprops=means_style,
    whiskerprops=whisker_style,
    flierprops=flier_style,
    medianprops=median_style,
    boxprops=box_style,
    notch=True,
)

plt.ylim(60, 100)
plt.tick_params(axis="both", direction="in", top=True, right=True, left=False)

plt.grid(color="black", linestyle="--", linewidth=0.33, dashes=(3, 20))
plt.xticks(rotation=30)

plt.savefig("myplot3.pdf")
plt.show()
```

```python
fig, (pltA, pltB) = plt.subplots(nrows=1, ncols=2, figsize=(8, 8))
fig.subplots_adjust(wspace=4)
# PLOT A
# Ustawienia markerów
top_axis_marker_ticks = slice(0, 200, 25)

# Ogólne ustawienia wszystkich plotów
mpl.rcParams["lines.markeredgecolor"] = "black"
mpl.rcParams["lines.markeredgewidth"] = 0.5
mpl.rcParams["lines.linewidth"] = 1

# Wykonujemy plot dla poszczególnych danych
pltA.plot(
    data_xaxis,
    data_val_avg[2],
    label=f"1-Evol-RS",
    color="blue",
    marker=".",
    markevery=top_axis_marker_ticks,
    markersize=10,
)
pltA.plot(
    data_xaxis,
    data_val_avg[1],
    label=f"1-Coev-RS",
    color="green",
    marker="v",
    markevery=top_axis_marker_ticks,
    markersize=7,
)
pltA.plot(
    data_xaxis,
    data_val_avg[4],
    label=f"2-Coev-RS",
    color="red",
    marker="d",
    markevery=top_axis_marker_ticks,
    markersize=6,
)
pltA.plot(
    data_xaxis,
    data_val_avg[0],
    label=f"1-Coev",
    color="black",
    marker="s",
    markevery=top_axis_marker_ticks,
    markersize=6,
)
pltA.plot(
    data_xaxis,
    data_val_avg[3],
    label=f"2-Coev",
    color="violet",
    marker="d",
    markevery=top_axis_marker_ticks,
    markersize=6,
)

# Ustawienia osi
pltA.set_xlabel(r"Rozegranych gier ($\times$1000)")
pltA.set_xlim(0, 500)
pltA.set_ylabel("Odsetek wygranych gier [%]")
pltA.set_ylim(60, 100)
pltA.tick_params(axis="both", direction="in", top=True, right=True, which="both")
pltA.grid(color="black", linestyle="--", linewidth=0.3, dashes=(3, 20))

# Ustawienia legendy
pltA.legend(loc="lower right", bbox_to_anchor=(1.0, 0.0), fontsize=12)
pltA.legend(numpoints=2, handlelength=2.0)

# Górna oś Pokoleń
top_axis = pltA.twiny()
top_axis.set_xlabel("Pokolenie")
top_axis.set_xlim(0, 200)
top_axis_ticks = np.arange(0, 201, 40)
top_axis.set_xticks(top_axis_ticks)
top_axis.tick_params(axis="x", direction="in")

# PLOT B

# Ustawienia dla poszczególnych parametrów boxplota
data_bp = [
    data_1ers[200, 2:] * 100,
    data_1crs[200, 2:] * 100,
    data_2crs[200, 2:] * 100,
    data_1c[200, 2:] * 100,
    data_2c[200, 2:] * 100,
]
flier_style = dict(
    marker="+",
    markerfacecolor="b",
    markeredgecolor="b",
    markeredgewidth=0.8,
    markersize=10,
)
median_style = dict(color="r", linewidth=2)
box_style = dict(color="b", linewidth=1.5)
whisker_style = dict(color="b", linewidth=1.5, linestyle="--", dashes=(8, 5.5))
means_style = dict(
    marker="o",
    markerfacecolor="b",
    markeredgecolor="black",
    markeredgewidth=2,
    markersize=6,
)
caps_style = dict(linewidth=1.5)

# Plotujemy nasze dane, włączamy wartośc 'mean' i ładujemy nasze parametry
pltB.boxplot(
    data_bp,
    labels=["1-Evol-RS", "1-Coev-RS", "2-Coev-RS", "1-Coev", "2-Coev"],
    capprops=caps_style,
    showmeans=True,
    meanprops=means_style,
    whiskerprops=whisker_style,
    flierprops=flier_style,
    medianprops=median_style,
    boxprops=box_style,
    notch=True,
)
pltB.set_ylim(60, 100)
pltB.tick_params(axis="both", direction="in", top=True, right=True)
pltB.tick_params(axis="x", rotation=30)
pltB.grid(color="black", linestyle="--", linewidth=0.33, dashes=(3, 20))

# Labels po prawej stronie
pltB.yaxis.tick_right()

plt.tight_layout()
plt.savefig("final_plot.pdf")
plt.show()
```
